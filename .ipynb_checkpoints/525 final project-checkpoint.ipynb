```python
categories = ['Nightlife', 'American(Traditional)', 'Sandwiches', 'FastFood', 'American(New)', 'Pizza', 'Mexican', 'Breakfast&Brunch', 'Burgers', 'Italian', 'Salad', 'Chinese', 'Seafood', 'ChickenWings', 'Coffee&Tea', 'EventPlanning&Services', 'Japanese', 'Delis', 'Steakhouses', 'AsianFusion', 'Barbeque', 'Desserts', 'Diners', 'Mediterranean',  'Bakeries', 'SpecialtyFood',  'Vegetarian', 'Soup', 'Thai',  'IceCream&FrozenYogurt', 'Tex-Mex', 'Buffets', 'JuiceBars&Smoothies', 'Gluten-Free']


```

```python
def cohend(d1, d2):
	# calculate the size of samples
	n1, n2 = len(d1), len(d2)
	# calculate the variance of the samples
	s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)
	# calculate the pooled standard deviation
	s = np.sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
	# calculate the means of the samples
	u1, u2 = np.mean(d1), np.mean(d2)
	# calculate the effect size
	return (u1 - u2) / s 

```

```python
category_dis = pd.DataFrame(columns=['business_id','cat', 'dis', 'review_count_z_dis'])

for category in categories:
    category_df = google_yelp[google_yelp[category]==1][['business_id', 'dis', 'review_count_z_dis']]
    category_df['cat'] = category
    category_dis = pd.concat([category_dis, category_df], axis = 0)
print(len(category_dis))
category_dis.head()



```

```python
#get cohen's d 
sig_categories = []
for category in categories:
    print(category)    
    print(ttest_ind(category_dis['dis'], category_dis[category_dis['cat']==category]['dis']))
    if (ttest_ind(category_dis['review_count_z_dis'], category_dis[category_dis['cat']==category]['review_count_z_dis'])[1]<0.05):
        sig_categories.append(category)
    print(cohend(category_dis['dis'], category_dis[category_dis['cat']==category]['dis']))
```

```python
category_dis_all = category_dis.copy()
category_dis_all = category_dis_all.reset_index(drop=True)
category_dis_all['cat'] = 'ALL'
category_dis_all = pd.concat([category_dis_all, category_dis[category_dis['cat'].isin(sig_categories)]], axis = 0)
```

```python
sns.set(rc={'figure.figsize':(16,9)}, font_scale = 1.5)
sns.boxplot(x="dis", y="cat", data=category_dis_all, palette="Set3")
plt.show()

#mod = ols('mean ~ chain',data=yelp[['mean', 'chain']]).fit()
mod = ols('dis ~ cat',data=category_dis).fit()                
aov_table = sm.stats.anova_lm(mod, typ=2)
aov_table.style.format('{0:,.2f}')
```

```python
sns.set(rc={'figure.figsize':(16,9)}, font_scale = 1.5)
sns.boxplot(x="review_count_z_dis", y="cat",  data=category_dis_all, palette="Set3")
plt.show()
#mod = ols('mean ~ chain',data=yelp[['mean', 'chain']]).fit()
mod = ols('review_count_z_dis ~ cat',data=category_dis).fit()                
aov_table = sm.stats.anova_lm(mod, typ=2)
aov_table.style.format('{0:,.2f}')
```
